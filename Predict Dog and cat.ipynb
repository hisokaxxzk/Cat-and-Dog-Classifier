```python
from tensorflow.keras.models import load_model
classifier = load_model("/content/drive/MyDrive/Colab Notebooks/Classify dog and cat CNN/dog_and_cat.h5")
```

```python
import numpy as np
link = '/content/drive/MyDrive/Colab Notebooks/Classify dog and cat CNN/single_prediction/cat_or_dog_9.jpg'
from tensorflow.keras.preprocessing import image
test_image = image.load_img(link,target_size = (64,64) )
img= image.load_img(link, target_size = (128,128))
test_image  = image.img_to_array(test_image)
test_image = test_image/255
test_image = np.expand_dims(test_image,axis = 0)
result = classifier.predict(test_image)
```
